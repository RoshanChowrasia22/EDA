### We will use logarithmic transformation

Why Log-Transform Skewed or Non-Normal Data in Regression?
In regression tasks (e.g., linear regression), models like ordinary least squares (OLS) rely on certain assumptions for reliable predictions and inference:

Linearity: The relationship between inputs (features) and output (target) should be linear.
Homoscedasticity: Constant variance of residuals (errors) across levels of the predictors.
Normality of residuals: Errors should be roughly normally distributed for valid statistical tests (e.g., p-values).

Skewed data (e.g., right-skewed like income, house prices, or exponential growth features) often violates these:

It introduces non-linearity (e.g., multiplicative effects look linear on a log scale).
It causes heteroscedasticity (variance increases with the variable's magnitude).
Residuals may not be normal, leading to biased estimates or poor model performance.

Why Take the Log of the Input Variable?
Applying a log transformation (e.g., log(x) or log(1 + x) for zero values) to skewed input features addresses these issues. Here's why it works:

Normalizes the Distribution:
Skewed data (e.g., exponential tails) becomes more symmetric and closer to normal after logging. This satisfies the normality assumption indirectly (by making residuals more normal).
Example: If your feature is "sales volume" (heavily right-skewed), log(sales) often looks bell-shaped.

Stabilizes Variance (Reduces Heteroscedasticity):
In skewed data, variance explodes at higher values. Logging compresses large values more than small ones, making variance more constant.
This leads to more reliable coefficient estimates and better prediction intervals.

Linearizes Relationships:
Many real-world processes are multiplicative (e.g., growth compounds exponentially), so log(y) = β₀ + β₁ log(x) + ε models percentage changes linearly (elasticities).
Without logging, the relationship might appear curved; with it, the model fits a straight line.

Improves Interpretability:
Coefficients become elasticities: A 1% change in the logged input leads to a β% change in the (logged) output.
Useful in domains like economics or biology.


Quick Example
Suppose you're predicting house prices (y) from size (x in sq ft), where x is right-skewed:

Raw model: y = β₀ + β₁ x + ε → Curved fit, uneven residuals.
Logged: y = β₀ + β₁ log(x) + ε → Straighter line, stable errors.

### Selection opf Best model

```python
from sklearn.model_selection import train_test_split,cross_val_score,GridSearchCV
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error,r2_score
import warnings
warnings.filterwarnings('ignore')
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_train,y_train,test_size=0.2,random_state=42)
```

```python
models = {
    'LinearRegression': {
        'model': LinearRegression(),
        'params': {}  # No hyperparameters to tune
    },
    'Lasso': {
        'model': Lasso(random_state=42),
        'params': {'alpha': [0.001, 0.01, 0.1, 1.0, 10.0]}
    },
    'Ridge': {
        'model': Ridge(random_state=42),
        'params': {'alpha': [0.001, 0.01, 0.1, 1.0, 10.0]}
    },
    'RandomForest': {
        'model': RandomForestRegressor(random_state=42),
        'params': {
            'n_estimators': [50, 100, 200],
            'max_depth': [None, 10, 20],
            'min_samples_split': [2, 5, 10]
        }
    },
    'AdaBoost': {
        'model': AdaBoostRegressor(random_state=42),
        'params': {
            'n_estimators': [50, 100, 200],
            'learning_rate': [0.01, 0.1, 1.0]
        }
    },
    'GradientBoost': {
        'model': GradientBoostingRegressor(random_state=42),
        'params': {
            'n_estimators': [50, 100, 200],
            'learning_rate': [0.01, 0.1, 0.2],
            'max_depth': [3, 5, 7]
        }
    }
}
```

### Hyperparameter tuning

```python
cv_folds=5
results={}
for name,config in models.items():
    print(name,config)
    model=config['model']
    params=config['params']

    if params:
        grid_search=GridSearchCV(model,params,cv=cv_folds,scoring='neg_mean_squared_error',n_jobs=-1,verbose=1)
        grid_search.fit(X_train,y_train)
        best_model=grid_search.best_estimator_
        best_score = grid_search.best_score_
    else:
        best_model=model
        best_model.fit(X_train,y_train)
        cv_scores=cross_val_score(model,X_train,y_train,cv=cv_folds,scoring='neg_mean_squared_error')
        best_score = np.mean(cv_scores)
    y_pred=best_model.predict(X_test)
    test_mse = mean_squared_error(y_test, y_pred)
    test_r2 = r2_score(y_test, y_pred)
    results[name] = {
        'best_model': best_model,
        'cv_neg_mse': best_score,
        'test_mse': test_mse,
        'test_r2': test_r2,
        'best_params': getattr(grid_search, 'best_params_', {})
    }
    
    print(f"Best CV neg MSE: {best_score:.4f}")
    print(f"Test MSE: {test_mse:.4f}, Test R²: {test_r2:.4f}")
    if params:
        print(f"Best params: {getattr(grid_search, 'best_params_', {})}")
```

```python
best_model_name = min(results.keys(), key=lambda k: results[k]['cv_neg_mse'])
best_model = results[best_model_name]
print(f"\nBest model based on CV: {best_model_name}")
print(f"CV neg MSE: {best_model['cv_neg_mse']:.4f}")
print(f"Test MSE: {best_model['test_mse']:.4f}, Test R²: {best_model['test_r2']:.4f}")
```
